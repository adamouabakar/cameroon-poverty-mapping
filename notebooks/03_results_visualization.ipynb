# Notebook 03 — Visualisation des résultats (v4)

Cartographie de la pauvreté au Cameroun : DHS 2018, GEE v3, INS ECAM 4.

```
from pathlib import Path
import json
from IPython.display import Image, display, Markdown

_cwd = Path.cwd()
PROJECT_ROOT = _cwd if (_cwd / 'outputs').exists() else _cwd.parent
REPORTS = PROJECT_ROOT / 'outputs/reports'
MAPS = PROJECT_ROOT / 'outputs/maps'
report = json.loads((REPORTS / 'model_v4_results.json').read_text())
m = report['metrics_v4_oof']
print(f"R² OOF v4: {m['r2']:.4f}")
print(f"Spearman v4: {m['spearman']:.4f}")
```

## Diagnostic modèle v4

```
plots = [
    'oof_scatter_v4.png', 'residuals_v4.png',
    'feature_importance_v4_top17.png', 'v3_vs_v4_metrics_viz.png',
]
for name in plots:
    p = REPORTS / name
    if p.exists():
        display(Markdown(f'### {name}'))
        display(Image(filename=str(p)))
```

## Cartes nationales v4

```
for name in [
    'wealth_index_predicted_1km_model_v4.png',
    'wealth_uncertainty_1km_model_v4.png',
    'priority_index_1km_v4.png',
    'wealth_national_clusters_v4.png',
]:
    p = MAPS / name
    if p.exists():
        display(Markdown(f'### {name}'))
        display(Image(filename=str(p)))
```

## Cartes régionales

```
for sub in ['regional', 'regional_v4']:
    reg_dir = MAPS / sub
    if reg_dir.exists():
        for p in sorted(reg_dir.glob('*.png')):
            display(Markdown(f'### {p.name}'))
            display(Image(filename=str(p)))
```
