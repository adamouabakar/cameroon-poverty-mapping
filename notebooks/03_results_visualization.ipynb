# Notebook 03 — Visualisation des résultats

Cartographie de la pauvreté au Cameroun (DHS 2018, features GEE v3).

```
from pathlib import Path
import json
import geopandas as gpd
import pandas as pd
from IPython.display import Image, display, Markdown

PROJECT_ROOT = Path('..').resolve()
REPORTS = PROJECT_ROOT / 'outputs/reports'
MAPS = PROJECT_ROOT / 'outputs/maps'
report = json.loads((REPORTS / 'real_model_results.json').read_text())
print('R² OOF:', report['metrics_oof']['r2'])
print('Spearman:', report['metrics_oof']['spearman'])
```

## Diagnostic modèle

```
for name in ['oof_scatter_viz.png', 'residuals_oof.png', 'feature_importance_top15.png', 'metrics_comparison.png']:
    p = REPORTS / name
    if p.exists():
        display(Markdown(f'### {name}'))
        display(Image(filename=str(p)))
```

## Cartes nationales et régionales

```
for name in ['wealth_national_clusters.png', 'uncertainty_national_clusters.png',
             'wealth_index_predicted_1km.png', 'wealth_uncertainty_1km.png']:
    p = MAPS / name
    if p.exists():
        display(Markdown(f'### {name}'))
        display(Image(filename=str(p)))
reg_dir = MAPS / 'regional'
if reg_dir.exists():
    for p in sorted(reg_dir.glob('*.png')):
        display(Markdown(f'### {p.name}'))
        display(Image(filename=str(p)))
```
